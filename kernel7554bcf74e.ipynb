The above profile makes the further objectives clear:
1. Churn for each kind of contract
2. Churn in case of customers with dependents and/or partners, and in case of senior citizens
3. Churn with respect to the kind of services availed - streaming movies or streaming tv
4. Relation between churn and tenure.
